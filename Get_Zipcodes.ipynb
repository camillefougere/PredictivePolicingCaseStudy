```
df = pd.read_csv('/gdrive/MyDrive/capstone/Cleaned_Crime_Data_2012.csv')
df.drop(columns='Unnamed: 0', inplace=True)
df['Date Occd'] = pd.to_datetime(df['Date Occd']).dt.normalize()
df['Date Rptd'] = pd.to_datetime(df['Date Rptd'])
df.sort_values(by='Date Rptd', inplace=True)
display(df.head(10))
```

```
focused = df[['Date Rptd', 'Part 1-2','COORDINATES']]
display(focused.head(1))
part1 = focused[focused['Part 1-2'] == 1]
display(part1.head(1))
part2 = focused[focused['Part 1-2'] == 2]
display(part2.head(1))
```

```
jan1 = part1[part1['Date Rptd'] < '2012-2-1']
jan1.reset_index(inplace=True)
print(jan1)
jan2 = part2[part2['Date Rptd'] < '2012-2-1']
jan2.reset_index(inplace=True)
print(jan2)
```

```
def get_zips(df):
  geolocator = Nominatim(user_agent="geoapiExercises")
  zips = []
  for i in range(len(df)):
    zip = geolocator.reverse(df.loc[i, 'COORDINATES']).raw['address'].get('postcode', 'missing')[:5]
    zips.append(zip)
    if len(zips)%500==0:
      print('Zips Added: ' + str(len(zips)))
  df.insert(len(df.columns), 'ZIPCODE', zips)
  return df
```

```
zipped_jan1 = get_zips(jan1)
```

```
zipped_jan1.to_csv('part1crimesjanuary.csv')
```

```
zipped_jan2 = get_zips(jan2)
```

```
zipped_jan2.to_csv('part2crimesjanuary.csv')
```
